```python
df = pd.concat([df, venue], axis = 1)
```

```python
features = df.drop(['total'], axis=1)
labels = df['total']
```

```python
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size=0.20, shuffle=True)
print(f"Training Set : {train_features.shape}\nTesting Set : {test_features.shape}")
```

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse
def evaluate_model(model):
    model.fit(train_features, train_labels)
    print('---------- model evaluation -----------')
    print('accuracy: ', model.score(test_features, test_labels)*100)
    print('mean absolute error: ', mae(test_labels, model.predict(test_features)))
    print('mean squared error: ', mse(test_labels, model.predict(test_features)))
    
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
evaluate_model(dtr)
```

```python
from sklearn.linear_model import LinearRegression
lrg = LinearRegression()
evaluate_model(lrg)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
evaluate_model(forest)
```

```python
from sklearn.svm import SVR
svr = SVR()
evaluate_model(svr)
```
